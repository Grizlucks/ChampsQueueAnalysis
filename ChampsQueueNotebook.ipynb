# A Quick Analysis Of LCS Players in Champion's Queue

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('CQdata.csv')
```

```python
clean = data[data.columns[1:8]].dropna()
```

```python
clean.dtypes
```

```python
clean.groupby('Name')['GP'].sum().sort_values().head(10).plot(kind='bar')
```

```python
clean.groupby('Name')['GP'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
```

```python
teams = ['EG', 'GG', 'C9', 'TL', 'FLY', 'CLG', 'DIG', '100', 'TSM', 'IMT']
pros = clean[clean['Team'].isin(teams)]
```

```python
pros.groupby('Name')['GP'].sum().sort_values().head(10).plot(kind='bar')
```

```python
pros.groupby('Name')['GP'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
```
